```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
ds = datasets.load_iris()
X = ds.data
y = ds.target
```

```python
X = ds.data[:,np.arange(2)]
```

```python
for output in range(0,3):
    X_iris = X[y == output,:]
    plt.scatter(X_iris[:,0],X_iris[:,1])
plt.show()
```
