```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,4*np.pi,1000)
C = np.cos(X)
S = np.sin(X)
```

```python
plt.figure()
plt.plot(X,C,'r')
plt.plot(X,S,'b')
plt.show()
```

```python
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
digit = 9
Xdig = X[y==digit]
np.random.shuffle(Xdig)
```

```python
plt.figure(figsize=(5,5))
for i in range(1,10):
    plt.subplot(3,3,i)
    plt.axis('off')
    plt.imshow(Xdig[i-1,:].reshape((8,8)),cmap='gray')
plt.show()
```
